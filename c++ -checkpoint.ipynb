# SWAPING

```python
#include<iostream>
using namespace std;
int main()
{
   # swap 
   int num1=5,num2=3;
   cout<<num1<<" "<<num2<<endl;
   int temp;
   
   temp=num1;
   num1=num2;
   num2=temp;
   
    cout<<num1<<" "<<num2<<endl;
}

'''5 3
   3 5'''
```
