```python
import numpy as np
x=np.linspace(0,10,10)
plt.figure(dpi=150)
plt.plot(x,x**1,label='x vs x', lw=3, ls='-',color='red',marker='o',markersize=9,markerfacecolor='yellow',markeredgewidth=3)
plt.plot(x,x**2,label='x vs x**2', lw=2, ls='--',color='blue',marker='^',markersize=3,markerfacecolor='yellow',markeredgewidth=3)
plt.plot(x,x**3,label='x vs x*3', lw=2, ls='-',color='green',marker='s',markersize=5,markerfacecolor='yellow',markeredgewidth=3)
plt.plot(x,x**4,label='x vs x*4', lw=4, ls='-',color='violet',marker='d',markersize=15,markerfacecolor='yellow',markeredgewidth=3)

plt.legend(loc=0)         # loc can take values from 0 to 10
plt.show()
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.15,0.5,0.4,0.4])

axes1.plot(x,x**2)
axes1.set_title('Outer Plot Title',fontsize=20)
axes1.set_xlabel('xlabel',fontsize=20,rotation=70)
axes1.set_ylabel('ylabel',fontsize=15,rotation=90)

axes2.plot(x,x**2)
axes2.set_title('Inner Plot Title',fontsize=17)
axes2.set_xlabel('xlabel',fontsize=15)
axes2.set_ylabel('ylabel',fontsize=15)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=4)

axes[0].plot(x,x**2)
axes[1].plot(x,x**2)
axes[2].plot(x,x**2)
axes[3].plot(x,x**2)

plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=4)
for element in axes:
    element.plot(x,x**2)
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=4)
for i in range(len(axes)):
    axes[i].plot(x,x**i+1)
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(nrows=3,ncols=4)
for i in range(3):
    for j in range(4):
        axes[i][j].plot(x,x**2)
        if(i==0):
            axes[i][j].plot(x,x**3)
        if(i==1):
            axes[i][j].plot(x,x**4)
        if(i==2):
            axes[i][j].plot(x,x*5)
plt.tight_layout()
plt.show()
```
