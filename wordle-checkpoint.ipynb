```python
import numpy as np
data =np.loadtxt("dictionary.txt",dtype='str')
data
```

```python
dataleft=[]
for i in data:
    if len(i)==6:
        dataleft.append(i)
dataleft
```

```python
def grey(char,dataleft):
    remaining=[]
    for i in dataleft:
        if char not in i:
            remaining.append(i)
    return remaining

grey('a',dataleft)
    
    
```

```python
def orange(char,pos,dataleft):
    remaining=[]
    for i in dataleft:
        if i[pos]!=char and (char in i):
            remaining.append(i)
    return remaining 
        
orange('b',1,dataleft)       
```

```python
def green(char,pos,dataleft):
    remaining=[]
    for i in dataleft:
        if i[pos]==char:
            remaining.append(i)
    return remaining

green('y',4,dataleft)
```

```python
import string
def letterpopularity(dataleft):
    alphabet=dict.fromkeys(string.ascii_lowercase,0)
    for i in dataleft:
        for j in i:
            alphabet[j]+=1
    return alphabet
letterpopularity(dataleft)    
```

```python
popularity=letterpopularity(dataleft)
def popularword(popularity,dataleft):
    max_sum=0
    mostpop=''
    for i in dataleft:
        sum=0
        for j in i:
            sum+=popularity[j]
        if sum>max_sum:
            mostpop=i
            max_sum=sum
    return mostpop

popularword(popularity,dataleft)
```

```python
dataleft=[]
for i in data:
    if len(i)==6:
        dataleft.append(i)
while True:
    ltpop=letterpopularity(dataleft)
    mostPopWord=popularword(ltpop,dataleft)
    print(mostPopWord)
    n=len(dataleft)
    if n==1:
        break
    response=input()
    for i in range(len(response)):
        if response[i]=='0':
            dataleft=grey(mostPopWord[i],dataleft)
        elif response[i]=='1':
            dataleft=orange(mostPopWord[i],i,dataleft)
        elif response[i]=='2':
            dataleft=green(mostPopWord[i],i,dataleft)
    
```
