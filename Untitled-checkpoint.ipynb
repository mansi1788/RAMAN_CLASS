```python
import numpy as np
import pandas as pd
```

```python
column_names=['movie id','movie title','release date ', 'video release date', 
              'IMDb URL' ,'unknown', 'Action' ,'Adventure' ,'Animation' ,
              'Childrens', 'Comedy','Crime','Documentary','Drama','Fantasy',
              'Film-Noir', 'Horror','Musical','Mystery','Romance','Sci-Fi',
              'Thriller', 'War','Western']
df=pd.read_csv('u.item',sep='|',names = column_names,encoding='latin-1')
df
              
```

```python
df.loc[4]['movie title']
```

# Create index - movie_title dictionary

```python
index_movie_dict={}
for i in range(len(df)):
    index_movie_dict[i]=df.loc[i]['movie title']
index_movie_dict
```

# Create movie_title-index dictionary

```python
movie_index_dict={}
for i in range(len(df)):
    movie_index_dict[df.loc[i]['movie title']]=i
movie_index_dict
```

```python
df=df.drop(['movie id','movie title','release date ', 'video release date', 
              'IMDb URL'],axis = 1)
df
```

# Create Similarity Matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(df)
similarity_matrix
```

```python
similarity_matrix.shape
```

```python
movie=str(input("Enter the recently watched movie: "))
number_of_recommendations=int(input("How many recommendations do you want?"))

idx = movie_index_dict[movie]
similarity_list = similarity_matrix[idx]
lst=[]
for i in range(len(similarity_list)):
    lst.append((similarity_list[i],i))
lst.sort(reverse = True)
recommendations=[]
for i in range(len(lst)):
    recommendations.append(index_movie_dict[lst[i][1]])
recommendations.remove(movie)
final_recommendation_list=recommendations[:number_of_recommendations]
final_recommendation_list
    
```
